```python
#2 Import libraries

import matplotlib.pyplot as plt #plotting
import numpy as np #scientific computing with arrays
import math #simple math like exp, pi, etc
```

```python
#3 Define functions

def boltzman_factor(m,g,h,kB,T):
    return math.exp(-(m*g*h)/(kB*T))

kB = 1.38064852*10**(-23) #m^2 kg s^(-2) K^(-1)
g = 9.807 # m/sÂ²
m = 5.32 * 10**(-23) *10**(-3) #kilo grams, weight of one O2 molecule
T = 293 #K

a=boltzman_factor(m,g,2,kB,T)
b=boltzman_factor(m,g,4,kB,T)
c=boltzman_factor(m,g,6,kB,T)
print(a,b,c)

boltzman_factor(m,g,6,kB,T)
```

```python
#5 Plotting (Way 1: Using functions)

plt.figure(dpi=100)
height = np.linspace(0,20000,100) #m
#print(height)

# blue plot
fraction = []
for h in height:
    fraction.append(boltzman_factor(m,g,h,kB,T+5))
plt.plot(height,fraction, '-g', label='apple')

# orange plot
fraction2 = []
for h in height:
    fraction2.append(boltzman_factor(2*m,g,h,kB,T))
plt.plot(height,fraction2, label='apple2')
plt.xlabel("height [m]")
plt.ylabel("Fraction of O2 molecules above height h")
plt.legend(loc='best')
```

```python
#6 Plotting (Way 2)

plt.figure(dpi=100)
h = np.linspace(0,20000,100) #m

f1= np.exp(-(m*g*h)/(kB*T))
f2= np.exp(-(2*m*g*h)/(kB*T))

plt.plot(h,f1)
plt.plot(h,f2)
plt.xlabel("height [m]")
plt.ylabel("Fraction of O2 molecules above height h")
```

```python
#7 Formatting and area under curve

def maxwell_boltzman_distribution(m,v,kB,T):
    normalizationFactor = math.sqrt(m/(2*math.pi*kB*T))
    return (normalizationFactor**3)*(4*math.pi*v**2)*(math.exp(-(m*v**2)/(2*kB*T)))

kB = 1.38064852*10**(-23) #m^2 kg s^(-2) K^(-1)
m = 5.32 * 10**(-23) *10**(-3) #kilo grams, weight of one O2 molecule
T1 = 293 #K 
T2 = 900 #K
T3 =1200

plt.figure(dpi=100)

v0 = 0
v1 = 2500
nSamples = 100000
dv = (v1-v0)/nSamples
velocities = np.linspace(v0,v1, nSamples) #m
frequency = []
for v in velocities:
    frequency.append(maxwell_boltzman_distribution(m,v,kB,T1))

plt.plot(velocities,frequency)

frequency2 = []
for v in velocities:
    frequency2.append(maxwell_boltzman_distribution(m,v,kB,T2))
    
plt.plot(velocities,frequency2)

frequency3 = []
for v in velocities:
    frequency3.append(maxwell_boltzman_distribution(m,v,kB,T3))
    
plt.plot(velocities,frequency3)

plt.xlabel("velocity [m/s]")
plt.ylabel("Probability density")

print("dv:{}".format(dv))
print("dv:",dv)
print("area under the curve: {}".format(np.sum(frequency)*dv))
```

```python
#8 Formatting and rounding figures

S_meth = 186 #J/(K mol) gas
S_o2 = 205 #J/(K mol) gas
S_co2 = 213.6 #J/(K mol) liquid
S_h20 = 69.9 # (J/K mol) gas

reactants = ( 1*S_meth + 2*S_o2 ) *10**(-3)
print("reactants: {} kJ/(K mol)".format(reactants))

products = ( 1*S_co2 + 2*S_h20 ) *10**(-3)
print("products: {} kJ/(K mol)".format(products))

deltaS = (products - reactants) 
print("deltaS: {} kJ/(K mol)".format(np.round(deltaS,decimals =4)))
```
